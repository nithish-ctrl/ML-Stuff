```python
df = pd.read_csv("https://github.com/ybifoundation/Dataset/raw/main/Salary%20Data.csv")
df.head()
```

```python
x=df[['Experience Years']]
```

```python
y=df['Salary']
```

```python
plt.scatter(x,y)
```

Linear Regression From Scratch in Python (Mathematical) : https://youtu.be/VmbA0pi2cRQ?si=0EhkE4wCsKjtGgz-

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=70106)
```

```python
x_train.shape,y_train.shape,y_test.shape,x_test.shape
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
c=model.intercept_
```

```python
m=model.coef_
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error
mean_absolute_error(y_test,y_pred)
```

```python
mean_absolute_percentage_error(y_test,y_pred)
```

Mean Squared Error.jpg

```python
mean_squared_error(y_test,y_pred)
```
